```python
 str1=input('輸入身高')
 str2=input('輸入體重')
 a=int(str1)
 b=a/100
 c=int(str2)
 bmi=c/(b*b)

if bmi<18.5:
  print("BMI:" ,bmi,"體位:過輕")
elif 18.5<=bmi<= 24:
  print("BMI:" ,bmi,"體位:適中")
elif 24<=bmi<27:
    print("BMI:" ,bmi,"體位:過重")
else :
   print("BMI:" ,bmi,"體位:肥胖")

```
